# Conclusion 
De las diferentes vacunas la sputnik es la mas extendida aunque no tiene muchas segundas dosis a comparacion de las otras vacunas que si bien tienen un numerob bajo de primeras dosis su % de segundas dosis aplicadas es mucho mayor, tambien podemos concluir que las mujeres fueron privilegiadas en cuanto a recibir una cantidad considerablemente mayor de dosis totales que los hombres. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
import matplotlib.pyplot as plt3
import matplotlib.pyplot as plt4
import numpy as np
df = pd.read_csv('dataset_total_vacunas(1).csv')
dfSputnik = df.query('VACUNA == "Sputnik"')
dfAstraZeneca = df.query('VACUNA == "AstraZeneca"')
dfSinopharm = df.query('VACUNA == "Sinopharm"')
dfModerna = df.query('VACUNA == "Moderna"')
#print(que vacunas se administraron mas cuantas primeras dosis que genero esta mas vacunado cuantas vacunas de cada tipo cuantas segundas dosis )
vacunasLen = [len(dfSputnik),len(dfAstraZeneca),len(dfSinopharm),len(dfModerna)]
vacunasName = ["sputnik", "AstraZeneca", "Sinopharm", "Morderna"]
plt.title('Numero de vacunas de cada tipo')
plt.xlabel('Nombre')
plt.ylabel('Cantidad')
plt.bar(vacunasName,vacunasLen)
plt.show()

primeraDosisPorVacuna = [dfSputnik["DOSIS_1"].sum(), dfAstraZeneca["DOSIS_1"].sum(), dfSinopharm["DOSIS_1"].sum(), dfModerna["DOSIS_1"].sum()]
segundaDosisPorVacuna = [dfSputnik["DOSIS_2"].sum(), dfAstraZeneca["DOSIS_2"].sum(), dfSinopharm["DOSIS_2"].sum(), dfModerna["DOSIS_2"].sum()]

print("")
plt.title('Numero de primeras dosis de cada tipo')
plt.xlabel('Nombre')
plt.ylabel('Cantidad')
plt.bar(vacunasName,primeraDosisPorVacuna)
plt.show()
print("")
plt.title('Numero de segundas dosis de cada tipo')
plt.xlabel('Nombre')
plt.ylabel('Cantidad')
plt.bar(vacunasName,segundaDosisPorVacuna)
plt.show()

dfSputnikH = dfSputnik.query('GENERO == "M"')
dfAstraZenecaH = dfAstraZeneca.query('GENERO == "M"')
dfSinopharmH = dfSinopharm.query('GENERO == "M"')
dfModernaH = dfModerna.query('GENERO == "M"')

dfSputnikF = dfSputnik.query('GENERO == "F"')
dfAstraZenecaF = dfAstraZeneca.query('GENERO == "F"')
dfSinopharmF = dfSinopharm.query('GENERO == "F"')
dfModernaF = dfModerna.query('GENERO == "F"')

primeraDosisPorVacunaHombres = dfSputnikH["DOSIS_1"].sum() + dfAstraZenecaH["DOSIS_1"].sum() + dfSinopharmH["DOSIS_1"].sum() + dfModernaH["DOSIS_1"].sum()
primeraDosisPorVacunaMujeres = dfSputnikF["DOSIS_1"].sum() + dfAstraZenecaF["DOSIS_1"].sum() + dfSinopharmF["DOSIS_1"].sum() + dfModernaF["DOSIS_1"].sum()

primeraDosisPorVacunaPorGenero = [primeraDosisPorVacunaHombres, primeraDosisPorVacunaMujeres]
generos = ["Hombres", "Mujeres"]

print("")
plt.title('Numero de primeras dosis por genero')
plt.xlabel('Genero')
plt.ylabel('Cantidad')
plt.bar(generos,primeraDosisPorVacunaPorGenero)
plt.show()


segundaDosisPorVacunaHombres = dfSputnikH["DOSIS_2"].sum() + dfAstraZenecaH["DOSIS_2"].sum() + dfSinopharmH["DOSIS_2"].sum() + dfModernaH["DOSIS_2"].sum()
segundaDosisPorVacunaMujeres = dfSputnikF["DOSIS_2"].sum() + dfAstraZenecaF["DOSIS_2"].sum() + dfSinopharmF["DOSIS_2"].sum() + dfModernaF["DOSIS_2"].sum()

segundaDosisPorVacunaPorGenero = [segundaDosisPorVacunaHombres, segundaDosisPorVacunaMujeres]

print("")
plt.title('Numero de segundas dosis por genero')
plt.xlabel('Genero')
plt.ylabel('Cantidad')
plt.bar(generos,segundaDosisPorVacunaPorGenero)
plt.show()

hombresSputnik = dfSputnikH["DOSIS_2"].sum() + dfSputnikH["DOSIS_1"].sum()
mujeresSputnik = dfSputnikF["DOSIS_2"].sum() + dfSputnikF["DOSIS_1"].sum()

numeroDeVacundasSputnikAplicadasACadaGenero = [hombresSputnik, mujeresSputnik]

print("")
plt.title('Numero de vacunas Sputnik aplicadas a cada genero')
plt.xlabel('Genero')
plt.ylabel('Cantidad')
plt.bar(generos,numeroDeVacundasSputnikAplicadasACadaGenero)
plt.show()
```
