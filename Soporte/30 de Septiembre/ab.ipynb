```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("csv.csv",sep=",",engine="python" )
df2 = df[["latitude","median_house_value","median_income"]]
grafico = sns.pairplot(df2,hue="latitude",palette="Spectral")
```

En el grafico de valor medio de las casas podemos apreciar que, cuanto mas grande la latitud el valor promedio aumenta significativamente

En el grafico de ingreso medio podemos apreciar como cuanto mayor sea la latitud la gente tiene la relacion mas baja de ingresos por valor de propiedad mientras que cuanto menor sea la latitud este valor se ve incrementado
