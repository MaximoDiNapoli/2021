```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

if __name__ == '__main__':

    pagina = requests.get("https://www.aoezone.net/")
    
    objeto_sopa = BeautifulSoup(pagina.content, 'html.parser')

    listaDeObjetos = list(objeto_sopa.children)

    entradas = objeto_sopa.find_all('div', {'class': 'porta-article-item'})

    dataRaw = []
    for i, entrada in enumerate(entradas):
        if not entrada.has_attr( "style" ):
            try:
                titulo = entrada.find('div', {'class': "block-container porta-article-container"}).find('a', {'class': "porta-article-header"}).find('div', {'class': "porta-header-text"}).find('span').getText().replace("  ", "")

                autor = entrada.find("div", {"class" : "block-body message-inner"}).find('a').getText()

                fecha = entrada.find('time', {'class': 'u-dt'}).getText().replace("  ", "")

                dataRaw.append([titulo.strip(), autor.strip(), fecha])
                print ("%d - %s  |  %s  |  %s" % (i + 1, titulo, autor, fecha))
            except:
                pass

    data = pd.DataFrame(dataRaw)
    data.to_csv("Torneos_AOE")



        
```
